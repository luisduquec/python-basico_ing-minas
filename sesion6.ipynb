# Graficas de Funciones

```python
def cuadratica(n):
    cuadrado = n**2
    return (cuadrado)

valores_x = list(range(6))
valores_y = []

for i in valores_x:
    cuadrado_i = cuadratica(i)
    valores_y.append(cuadrado_i)
```

```python
print(valores_x)
print(valores_y)
```

```python
plt.plot(valores_x, valores_y)
```

```python
valores_x = list(range(11))
valores_y2 = list(map(cuadratica,valores_x))

plt.plot(valores_x,valores_y2)
```

# Graficas implementando la libreria Numpy

```python
import numpy as np
```

```python
x = np.linspace(0,5)
y = list(map(cuadratica,x))

x_p = [1,2,3,4,5]
y_p = list(map(cuadratica,x_p))

plt.xlabel("x")
plt.ylabel("y = x^2")
plt.title("Grafico de la funcion cuadratica")
plt.axis([1,4,0,16])

plt.plot(x,y, label = "Valores teoricos")
plt.plot(x_p,y_p,"or", label ="Valores observados")
plt.legend()
plt.show()
```
